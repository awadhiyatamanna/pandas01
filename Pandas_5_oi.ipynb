## Dropping row/columns where NaN is present

```python
a = np.random.randint(1,50,40).reshape(10,4)
df = pd.DataFrame(a,columns=list("ABCD"))
df
```

```python
df.loc[1,"B"] = np.nan
df.loc[1,"D"] = np.nan
df.loc[4,"D"] = np.nan
df
```

```python
df.dropna() # by default the row where NaN is present will be dropped 
```

```python
df.dropna(axis=1)
```
